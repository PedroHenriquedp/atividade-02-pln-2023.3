Um `programa Python` que utilize `expressões regulares` pode ajudar a identificar vários **tipos de erros** comuns em **livros**, especialmente erros de formatação e problemas relacionados à consistência do texto. Aqui estão alguns exemplos de erros comuns que podem ser identificados usando expressões regulares:

* Erros de gramática e ortografia: erros de digitação, concordância verbal e nominal, uso incorreto de pontuação e outros erros gramaticais.

* Problemas de formatação: você pode usar expressões regulares para encontrar erros de formatação, como espaços em excesso, tabulações inadequadas ou alinhamentos inconsistentes.

* Abreviações e acrônimos: você pode usar expressões regulares para encontrar abreviações ou acrônimos que não foram definidos ou explicados anteriormente no texto.

* Citações e referências: expressões regulares podem ser úteis para localizar citações ou referências que precisam de formatação especial.

* OUTROS TIPOS DE ERROS: não considerem apenas os tipos de erros citados acima.


**IMPORTANTE:** Lembre-se de que expressões regulares podem ser poderosas, mas também complexas. Dependendo da complexidade dos erros que você deseja identificar, pode ser necessário ajustar as expressões regulares de acordo com as características específicas do seu texto. Além disso, é importante ter em mente que as expressões regulares podem não ser a melhor ferramenta para todos os tipos de erros em livros, especialmente problemas mais contextuais ou semânticos, que podem exigir abordagens de PLN mais avançadas.

```
import requests
from bs4 import BeautifulSoup
import random
import re


class WebBookErrorDetector:

    def __init__(self, url):
        self.url = url
        self.requestData = None
        self.htmlContent = None
        self.contentPage = None
        self.errors = []
        return

    def process(self):
        print("\nSolicitando url: " + self.url)
        self._getData()
        self._getMainContent()
        self._findErrors()
        self._findErrorsHTML()
        self._generateReport()
        return

    def _getData(self):
        user_agents = [
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36",
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Edge/16.16299",
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Firefox/58.0",
            "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/602.4.8 (KHTML, like Gecko) Version/10.0.3 Safari/602.4.8",
            "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36",
        ]
        headers = {'User-Agent': random.choice(user_agents)}
        data = requests.get(self.url, headers=headers)

        if data.status_code != 200:
            print("Error: Não foi possível acessar a página")
            exit(1)

        self.requestData = data

    def _getMainContent(self):
        soup = BeautifulSoup(self.requestData.content, 'html.parser')
        self.htmlContent = soup.prettify()
        self.contentPage = (soup.find("main", {"class": "content"})).text
        return

    def _findErrors(self):
        #expressões para encontrar erros básicos de português pt-Br no texto
        expressions_regex = [
            {
                "name": "Espaço antes de ponto",
                "regex": "\s\."
            },
            {
                "name": "Espaço antes de vírgula",
                "regex": "\s\,"
            },
            {
                "name": "Url finalizada com ponto",
                "regex": "https?:.*\.↩︎?"
            },
            {
                "name": "Espaço antes de dois pontos",
                "regex": "\s\:"
            },
            {
                "name": "Espaço antes de ponto e vírgula",
                "regex": "\s\;"
            },
            {
                "name": "Espaço antes de exclamação",
                "regex": "\s\!"
            },
            {
                "name": "Espaço antes de interrogação",
                "regex": "\s\?"
            },
            {
                "name": "Letra minúscula depois de ponto final",
                "regex": "(?<!\bp|ex)(?<!\be.g)[.!?](?:\s|\n)+[a-z]"
            },
            {
                "name": "Letra minúscula depois de ponto de exclamação",
                "regex": "\!\s[a-z]"
            },
            {
                "name": "Letra minúscula depois de ponto de interrogação",
                "regex": "\?\s[a-z]"
            },
            {
                "name": "Letra minúscula depois de ponto final e quebra de linha",
                "regex": "\.\n[a-z]"
            },
            {
                "name": "Letra minúscula depois de ponto de exclamação",
                "regex": "\!\n[a-z]"
            },
            {
                "name": "Letra minúscula depois de ponto de interrogação",
                "regex": "\?\n[a-z]"
            },
            {
                "name": "Palavras repetidas em sequência",
                "regex": "\b(\w+)\s+\1\b"
            },
            {
                "name": "Letra minúscula após quebra de linha",
                "regex": "\n[a-z]"
            },
            {
                "name": "Travessão no início da linha seguido de espaço e letra minúscula",
                "regex": "\n-\s[a-z]"
            },
            {
                "name": "Letras 'a' e 'o' separadas",
                "regex": "\s[Aa]\s[Oo]\s"
            },
            {
                "name": "Sem Virgula antes de 'etc'",
                "regex": "[^,]\setc"
            },
            {
                "name": "Ponto depois de 'etc'",
                "regex": "etc[^.]"
            },
            {
                "name": "Espaço duplo",
                "regex": "  +"
            },
            {
                "name": "Carácter invisível",
                "regex": "[ ]+"
            },
            {
                "name": "Não tem espaço depois da vírgula",
                "regex": ",[^ ][a-z]"
            },
            {
                "name": "Não tem espaço depois de dois-pontos",
                "regex": "[:][^ ][a-z|0-9]"
            },
            {
                "name": "Não tem espaço depois de exclamação",
                "regex": "[!][^ ][a-z]"
            },
            {
                "name": "Não tem espaço depois de ponto e vírgula",
                "regex": "[;][^ ][a-z]"
            },
            {
                "name": "Não tem espaço depois de interrogação e tem letra minúscula depois",
                "regex": "[?][^ ][a-z]"
            },
            {
                "name": "Espaço depois do abre parenteses",
                "regex": "\(\s+"
            },
            {
                "name": "Espaço antes do fecha parenteses",
                "regex": "\s+\)"
            },

        ]
        #Loop para encontrar erros no texto
        for expression in expressions_regex:
            regex = re.compile(expression["regex"])
            matches = regex.findall(self.contentPage)
            self.errors.append({
                "name": expression["name"],
                "matches": matches
            })

        return

    def _findErrorsHTML(self):
        #expressões para encontrar erros no html
        expressions_regex = []

        #Loop para encontrar erros no html
        for expression in expressions_regex:
            regex = re.compile(expression["regex"])
            matches = regex.findall(self.htmlContent)
            self.errors.append({
                "name": expression["name"],
                "matches": matches
            })

    def _generateReport(self):
        print("Relatório de erros encontrados:")
        count_errors = 0
        for error in self.errors:
            print("-> " + error["name"])
            print("-- Quantidade: " + str(len(error["matches"])))
            count_errors += len(error["matches"])

        print("* Total de erros encontrados: " + str(count_errors))


if __name__ == '__main__':
    print("Iniciando análise ...")
    urls = [
        "https://brasileiraspln.com/livro-pln/1a-edicao/parte1/cap1/cap1.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte2/cap2/cap2.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte2/cap3/cap3.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte3/cap4/cap4.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte3/cap5/cap5.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte4/cap6/cap6.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte4/cap7/cap7.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap8/cap8.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap9/cap9.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte5/cap10/cap10.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap11/cap11.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap12/cap12.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap13/cap13.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte7/cap14/cap14.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte7/cap15/cap15.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap16/cap16.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap17/cap17.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap18/cap18.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap19/cap19.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte8/cap20/cap20.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte9/cap21/cap21.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte9/cap22/cap22.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte9/cap23/cap23.html",
        "https://brasileiraspln.com/livro-pln/1a-edicao/parte10/cap24/cap24.html",
        #"https://brasileiraspln.com/livro-pln/1a-edicao/parte10/cap25/cap25.html"
    ]

    for url in urls:
        pr = WebBookErrorDetector(url)
        pr.process()

    print("Análise finalizada!")
```
